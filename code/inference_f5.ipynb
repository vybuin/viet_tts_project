## 3. Load Model

```python
# Run with flags
# Leave --ref_text "" will have ASR model transcribe (extra GPU memory usage)
# !f5-tts_infer-cli --model F5TTS_v1_Base \
# --ref_audio REF_AUDIO \
# --ref_text REF_TEXT \
# --gen_text GEN_TEXT
```

```python
from f5_tts.api import F5TTS

tts = F5TTS(
    model="F5TTS_v1_Base",
    ckpt_file=CHECKPOINT,
    vocab_file=VOCAB_FILE,
    device="cuda",
    # use_ema=True,
)
print(f"Loaded checkpoint: {CHECKPOINT}")
```

## 4. Generate Speech

```python
# !unzip -q /content/mdv-tts/F5-TTS/data/1hr_data_subset.zip -d /content/mdv-tts/F5-TTS/data/
# !unzip -q /content/mdv-tts/F5-TTS/data/viet_tts_project_data.zip -d /content/
```

```python
import soundfile as sf
from IPython.display import Audio, display, Markdown

wav, sr, _ = tts.infer(
    ref_file=SOUTH_REF,
    ref_text=SOUTH_TEXT,
    gen_text=GEN_TEXT,
    # gen_text="Chúc mừng năm mới.",
)

output_path = os.path.join(OUTPUT_DIR, "generated.wav")
sf.write(output_path, wav, sr)

display(Markdown(f"**Generated text:** {GEN_TEXT}"))
display(Markdown(f"**Saved to:** `{output_path}`"))
display(Audio(output_path))

wav, sr, _ = tts.infer(
    ref_file=CENTRAL_REF,
    ref_text=Central_TEXT,
    gen_text=GEN_TEXT,
    # gen_text="Chúc mừng năm mới.",
)

output_path = os.path.join(OUTPUT_DIR, "generated.wav")
sf.write(output_path, wav, sr)

display(Markdown(f"**Generated text:** {GEN_TEXT}"))
display(Markdown(f"**Saved to:** `{output_path}`"))
display(Audio(output_path))

wav, sr, _ = tts.infer(
    ref_file=NORTH_REF,
    ref_text=North_TEXT,
    gen_text=GEN_TEXT,
    # gen_text="Chúc mừng năm mới.",
)

output_path = os.path.join(OUTPUT_DIR, "generated.wav")
sf.write(output_path, wav, sr)

display(Markdown(f"**Generated text:** {GEN_TEXT}"))
display(Markdown(f"**Saved to:** `{output_path}`"))
display(Audio(output_path))
```
